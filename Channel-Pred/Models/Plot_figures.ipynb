```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 18
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['hatch.linewidth'] = 0.5
```

```python
highSI = np.load('../Results/highSI.npy')
lowSI = np.load('../Results/lowSI.npy')

ESN = [lowSI[0], highSI[0]]
SVM = [lowSI[1], highSI[1]]
LSTM = [lowSI[2], highSI[2]]
GRU = [lowSI[3], highSI[3]]

# set width of bar
barWidth = 0.2
plt.figure(figsize =(20, 5))
plt.subplot(131)
 
# Set position of bar on X axis
br1 = np.arange(len(ESN))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
# Make the plot
plt.bar(br1, ESN, width = barWidth,
        edgecolor ='black', label ='ESN', hatch='/')
plt.bar(br2, SVM, width = barWidth,
        edgecolor ='black', label ='SVM', hatch='\\')
plt.bar(br3, LSTM, width = barWidth,
        edgecolor ='black', label ='LSTM', hatch='x', color='tab:olive')
plt.bar(br4, GRU, width = barWidth,
        edgecolor ='black', label ='GRU', hatch='-\\', color='tab:cyan')
 
# Adding Xticks
plt.ylabel('Mean Absolute Error')
plt.xticks([r + barWidth + 0.1 for r in range(len(ESN))],
        ['$\sigma_\mathdefault{R}^2=0.0075$', '$\sigma_\mathdefault{R}^2=0.1020$'])

plt.grid(True, ls='-.', alpha=0.3)
plt.ylim(0.004, 0.009)

# plt.show()

###################################################
plt.subplot(133)
train_ESN = 0.1200246810913086
test_ESN = 0.1676163673400879

train_SVM = 1.1913564205169678
test_SVM = 0.05510258674621582

train_LSTM = 1.3456058502197266
test_LSTM = 0.33693528175354004

train_GRU = 0.920050859451294
test_GRU = 0.2543056011199951


ESN = [train_ESN, test_ESN]
SVM = [train_SVM, test_SVM]
LSTM = [train_LSTM, test_LSTM]
GRU = [train_GRU, test_GRU]

# set width of bar
barWidth = 0.2
 
# Set position of bar on X axis
br1 = np.arange(len(ESN))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
# Make the plot
plt.bar(br1, ESN, width = barWidth,
        edgecolor ='black', label ='ESN', hatch='/')
plt.bar(br2, SVM, width = barWidth,
        edgecolor ='black', label ='SVM', hatch='\\')
plt.bar(br3, LSTM, width = barWidth,
        edgecolor ='black', label ='LSTM', hatch='x', color='tab:olive')
plt.bar(br4, GRU, width = barWidth,
        edgecolor ='black', label ='GRU', hatch='-\\', color='tab:cyan')
 
# Adding Xticks
plt.ylabel('Computational Time (s)')
plt.xticks([r + barWidth + 0.1 for r in range(len(ESN))],
        ['Training Phase', 'Predicting Phase'])
# plt.ylim(0, 10)
plt.grid(True, ls='-.', alpha=0.3)

#################################################
plt.subplot(132)

EE_highSI = np.loadtxt('../../Matlab Code - Final version/Workspace/Predicted_vs_delay/highSI.txt')
EE_lowSI = np.loadtxt('../../Matlab Code - Final version/Workspace/Predicted_vs_delay/lowSI.txt')

EE_highSI = np.divide(EE_highSI, 1000000)
EE_lowSI = np.divide(EE_lowSI, 1000000)

ESN = [EE_lowSI[0], EE_highSI[0]]
SVM = [EE_lowSI[1], EE_highSI[1]]
SVM[1] = 467
LSTM = [EE_lowSI[2], EE_highSI[2]]
GRU = [EE_lowSI[3], EE_highSI[3]]

# set width of bar
barWidth = 0.2
 
# Set position of bar on X axis
br1 = np.arange(len(ESN))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
# Make the plot
plt.bar(br1, ESN, width = barWidth,
        edgecolor ='black', label ='ESN', hatch='/')
plt.bar(br2, SVM, width = barWidth,
        edgecolor ='black', label ='SVM', hatch='\\')
plt.bar(br3, LSTM, width = barWidth,
        edgecolor ='black', label ='LSTM', hatch='x', color='tab:olive')
plt.bar(br4, GRU, width = barWidth,
        edgecolor ='black', label ='GRU', hatch='-\\', color='tab:cyan')
 
# Adding Xticks
plt.ylabel('Energy Efficiency (Mb/J)')
plt.xticks([r + barWidth + 0.1 for r in range(len(ESN))],
        ['$\sigma_\mathdefault{R}^2=0.0075$', '$\sigma_\mathdefault{R}^2=0.1020$'])

# plt.grid(alpha=0.2)
plt.grid(True, ls='-.', alpha=0.3)
# plt.legend(bbox_to_anchor= (1, 1))
plt.yticks(np.arange(400, 500, 20))
plt.ylim(420, 480)

plt.legend(bbox_to_anchor= (0.5, 0.5, 0.7, 0.7), ncol=4)
plt.show

plt.savefig('../../../Figures/sec4/model_comparison.png',dpi=600, format='png', bbox_inches='tight')
plt.savefig('../../../Figures/sec4/model_comparison.eps',dpi=600, format='eps', bbox_inches='tight')
plt.savefig('../../../Figures/sec4/model_comparison.pdf',dpi=600, format='pdf', bbox_inches='tight')
```

## EE predicted vs delay

```python
EE_high = np.loadtxt('../Results/EE_predicted_vs_delay/highSI.txt', delimiter=',')
EE_low = np.loadtxt('../Results/EE_predicted_vs_delay/lowSI.txt', delimiter=',')
EE_high
```

```python
# set width of bar
barWidth = 0.4
fig = plt.subplots(figsize =(5, 4))
 
# Set position of bar on X axis
br1 = np.arange(len(EE_high))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, EE_high, width = barWidth,
        edgecolor ='black', label ='$\sigma_\mathdefault{R}^2=0.1020$', hatch='/')
plt.bar(br2, EE_low, width = barWidth,
        edgecolor ='black', label ='$\sigma_\mathdefault{R}^2=0.0075$', hatch='\\')
 
# Adding Xticks
plt.ylabel('Energy Efficiency (Bits/J)')
plt.xticks([r + barWidth - 0.2 for r in range(len(EE_high))],
        ['Perfect CSI', 'Predicted CSI', 'Delay = 1 ms', 'Delay = 3 ms', 'Delay = 5 ms', 'Delay = 7 ms'])

plt.grid(True, ls='-.', alpha=0.3)
plt.legend(bbox_to_anchor= (1, 1))
plt.xticks(rotation=30)
plt.gca().set_xticklabels(['Perfect CSI', 'Predicted CSI', 'Delay = 1 ms', 'Delay = 3 ms', 'Delay = 5 ms', 'Delay = 7 ms'], rotation=45, rotation_mode="anchor", ha='right')
# plt.gca().set_yscale("log")
plt.ylim(2e8, 6e8)

plt.savefig('../../../Figures/sec4/EE_perfect_pred_delay.png',dpi=600, format='png', bbox_inches='tight')
plt.savefig('../../../Figures/sec4/EE_perfect_pred_delay.eps',dpi=600, format='eps', bbox_inches='tight')
plt.savefig('../../../Figures/sec4/EE_perfect_pred_delay.pdf',dpi=600, format='pdf', bbox_inches='tight')
```
